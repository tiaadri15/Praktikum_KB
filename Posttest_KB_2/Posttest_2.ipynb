### Menampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka. Kemudian dimasukkan dalam bentuk dictionary dengan key sesuai dengan nama informasinya

```python
#Menampilkan jumlah baris atau record
jumlah_record = len(carPrice)
#Menampilkan jumlah kolom atau attribute
jumlah_attribute = len (carPrice.columns)
#Menampilkan nama attribute
nama_attribute = carPrice.columns.tolist()
#Menampilkan jumlah attribute bertipe angka
attribute_tipe_angka = len (carPrice.select_dtypes(include = ["number"]).columns)

#Dictionary
informasi_dict_carPrice = {
    "Jumlah Record" : jumlah_record,
    "Jumlah Attribute" : jumlah_attribute,
    "Nama Attribute" : nama_attribute,
    "Jumlah Attribute bertipe Angka" : attribute_tipe_angka
}

#Menampilkan nama attribute agar hasilnya tidak ada kurung siku
nama_atribut_tanpa_kurung_siku = ", ".join(informasi_dict_carPrice["Nama Attribute"])

for key, value in informasi_dict_carPrice.items():
    if key == "Nama Attribute":
        print(f"{key} :")
        for attribute in nama_attribute:
            print(f"- {attribute}")
    else:
        print(f"{key} : {value}")
```

### Menampilkan jumlah nilai dari semua attribute yang bernilai object 

```python
# Menggunakan loop untuk menghitung jumlah nilai dari semua atribut bertipe data objek
for kolom in carPrice.select_dtypes(include=['object']):
    nilai_counts = carPrice[kolom].value_counts()
    print(f"Jumlah nilai untuk atribut '{kolom}':\n{nilai_counts}\n")
```
